```python
import subprocess
from ast import literal_eval

def run(command):
    process = subprocess.Popen(command, shell=True, stdout=subprocess.PIPE)
    out, err = process.communicate()
    print(out.decode('utf-8').strip())
```

```python
print('# CPU')
run('cat /proc/cpuinfo | egrep -m 1 "^model name"')
run('cat /proc/cpuinfo | egrep -m 1 "^cpu MHz"')
run('cat /proc/cpuinfo | egrep -m 1 "^cpu cores"')
```

```python
print('# RAM')
#run('cat /proc/meminfo | egrep "^MemTotal"')
import subprocess

def get_mem_total():
    try:
        output = subprocess.check_output("cat /proc/meminfo | grep '^MemTotal'", shell=True)
        output = output.decode('utf-8').strip()
        mem_total_kb = int(output.split()[1])
        mem_total_gb = mem_total_kb / (1024 * 1024)
        return round(mem_total_gb, 2)
    except Exception as e:
        print(f"An error occurred: {e}")
        return None

mem_total_gb = get_mem_total()
if mem_total_gb is not None:
    print(f"Total Memory: {mem_total_gb} GB")
```
